```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Läs in JSON-filen
with open("analysis_results.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
# Samla alla labels
all_labels = []

for image in data:
    for label in image["labels"]:
        all_labels.append(label["name"])
```

```python
# Räkna hur många gånger varje label förekommer
label_counts = Counter(all_labels)
print(label_counts)
```

```python
# Gör om till DataFrame
df = pd.DataFrame(label_counts.items(), columns=["Label", "Antal"])
df = df.sort_values("Antal", ascending=False)
print(df)
```

Eftersom det är otoligt många olika labels i mina bilder väljer jag att ta ut en top 20 av dem vanligaste förekommande, annars hade det blivit väldigt grötigt för visualiseringen

```python
top_20 = df.head(20)
print(top_20)
```

```python
plt.figure(figsize=(10,6))
plt.bar(top_20["Label"], top_20["Antal"])
plt.title("20 vanligaste förekommande labels i bilderna")
plt.xlabel("Label")
plt.ylabel("Antal")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
